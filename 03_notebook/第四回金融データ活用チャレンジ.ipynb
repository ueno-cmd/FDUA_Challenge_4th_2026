```python
# ===== Step 2: ドキュメント検索関数 =====

def search_documents(query, top_k=5):
    """
    クエリに関連する有価証券報告書のチャンクを検索
    
    Parameters:
    - query: 検索クエリ（日本語可）
    - top_k: 取得する上位N件（デフォルト5件）
    
    Returns:
    - pandas DataFrame: 関連チャンク
    """
    
    # キーワードベースの検索
    results = spark.sql(f"""
        SELECT 
            chunk_id,
            chunk_content,
            chunk_start_page,
            chunk_end_page,
            path
        FROM workspace.agent_workshop.chunked_documents
        WHERE chunk_content LIKE '%{query}%'
        ORDER BY chunk_start_page
        LIMIT {top_k}
    """)
    
    df = results.toPandas()
    print(f"検索完了: {len(df)}件のチャンクを取得")
    return df

# テスト実行
test_query = "成長戦略"
test_results = search_documents(test_query, top_k=3)

print(f"\n検索クエリ: {test_query}")
print("\n取得したチャンク:")
display(test_results[['chunk_id', 'chunk_start_page', 'chunk_content']])
```

```python
def search_documents_by_code(company_code, top_k=5):
    """
    企業コードで有価証券報告書チャンクを検索
    Parameters:
    - company_code: 企業コード（intまたはstr）
    - top_k: 取得する上位N件（デフォルト5件）
    Returns:
    - pandas DataFrame: 関連チャンク
    """
    # path列に企業コードが含まれるチャンクのみ抽出
    results = spark.sql(f"""
        SELECT 
            chunk_id,
            chunk_content,
            chunk_start_page,
            chunk_end_page,
            path
        FROM workspace.agent_workshop.chunked_documents
        WHERE path LIKE '%{company_code}%'
        ORDER BY chunk_start_page
        LIMIT {top_k}
    """)
    df = results.toPandas()
    print(f"検索完了: {len(df)}件のチャンクを取得")
    return df

# テスト実行
# 例: 12044（データに存在する企業コード）
test_company_code = 12044
test_results = search_documents_by_code(test_company_code, top_k=3)

print(f"\n検索企業コード: {test_company_code}")
print("\n取得したチャンク:")
display(test_results[["chunk_id", "chunk_start_page", "chunk_content"]])
```

## Step 3: 提案書生成関数

```python
from mlflow.deployments import get_deploy_client

# Databricks Serving Endpointクライアント初期化
client = get_deploy_client("databricks")

def generate_proposal(company_code, top_k=5):
    """
    財務データと有価証券報告書を基に投資提案書を生成
    Parameters:
    - company_code: 対象企業コード
    - top_k: 取得するチャンク数
    Returns:
    - str: 生成された提案書
    """
    print(f"=== 提案書生成開始: {company_code} ===\n")
    # 1. 財務データ取得
    print("1. 財務データを取得中...")
    company_financial = financial_df[financial_df['コード'] == company_code]
    if company_financial.empty:
        print(f"警告: {company_code}の財務データが見つかりません")
        financial_context = "財務データなし"
    else:
        financial_context = company_financial.to_string(index=False)
        print(f"   ✓ {len(company_financial)}行の財務データを取得")
    # 2. 有価証券報告書から関連情報取得（企業コード検索）
    print(f"\n2. 有価証券報告書を企業コードで検索中...")
    doc_chunks = search_documents_by_code(company_code, top_k=top_k)
    if doc_chunks.empty:
        print("警告: 関連する報告書が見つかりません")
        doc_context = "関連する報告書なし"
    else:
        doc_context = "\n\n".join([
            f"[ページ {row['chunk_start_page']}-{row['chunk_end_page']}]\n{row['chunk_content']}"
            for _, row in doc_chunks.iterrows()
        ])
        print(f"   ✓ {len(doc_chunks)}件のチャンクを取得")
    # 3. プロンプト構築
    print("\n3. プロンプトを構築中...")
    system_prompt = """あなたは経験豊富な投資アナリストです。
企業の財務データと有価証券報告書を分析し、投資家向けの提案書を作成してください。

提案書には以下を含めてください：
1. エグゼクティブサマリー
2. 企業概要と事業内容
3. 財務分析（収益性、成長性、安定性）
4. 強みと機会
5. リスクと課題
6. 投資推奨（買い/中立/売り）とその根拠

プロフェッショナルで説得力のある文章を心がけてください。"""
    user_prompt = f"""
## 対象企業コード
{company_code}

## 財務データ
{financial_context}

## 有価証券報告書（抜粋）
{doc_context}

上記の情報を基に、投資提案書を作成してください。
"""
    # 4. Azure OpenAI API（GPT-5）で生成
    print("\n4. Azure OpenAI API (GPT-5) で提案書を生成中...")
    response = client.predict(
        endpoint="openai-azu-gpt5-chat",
        inputs={
            "messages": [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt}
            ],
            "max_tokens": 4000,
            "temperature": 0.7
        }
    )
    proposal = response['choices'][0]['message']['content']
    print("\n=== 提案書生成完了 ===\n")
    return proposal

# テスト実行
# 例: 12044（データに存在する企業コード）
test_company_code = 12044

test_proposal = generate_proposal(test_company_code, top_k=3)

print("=" * 80)
print(test_proposal)
print("=" * 80)
```

## Step 4: 提案書評価関数

```python
# Step 4: 提案書評価関数

def evaluate_proposal(proposal_text, evaluation_prompt):
    """
    提案書を評価基準に従ってAIで評価
    Parameters:
    - proposal_text: 提案書本文（str）
    - evaluation_prompt: 評価プロンプト（str）
    Returns:
    - str: 評価結果
    """
    from mlflow.deployments import get_deploy_client
    client = get_deploy_client("databricks")
    # プロンプト構築
    full_prompt = f"""
{evaluation_prompt}

# 提案書
{proposal_text}
"""
    response = client.predict(
        endpoint="openai-azu-gpt5-chat",
        inputs={
            "messages": [
                {"role": "user", "content": full_prompt}
            ],
            "max_tokens": 2000,
            "temperature": 0.3
        }
    )
    return response['choices'][0]['message']['content']

# テスト実行
# 評価プロンプト（ユーザー依頼内容）
evaluation_prompt = '''# 依頼内容
あなたは大手銀行の支店長、および熟練の経営コンサルタントです。
現在開催されている「建設業向け事業提案コンペティション」において、提案書を作成してください。

# 評価基準
以下のポイントで作成してください。

| 評価視点 | 評価項目 | 評価のポイント（高評価の基準） |
| :--- | :--- | :--- |
| 1. 全体構成 | 過去分析と未来提案の接続 | 過去3年の財務・事業分析（Past）と、未来への成長戦略（Future）が一貫した因果関係で論理的に接続されているか。 |
| 2. 地域性 | 地域特性の考慮 | 企業の所在地（商圏、人口動態、行政施策）を踏まえ、画一的ではない地域密着型の提案ができているか。 |
| 3. 業界特性 | 販路・商流の理解 | 官公庁／民間、元請／下請などの販路特性を把握し、それに応じた具体的な分析・提案となっているか。 |
| 4. 業界課題1 | 近未来への対応 (GX/DX) | 低コスト工法、環境技術（GX）、省力化技術（DX）など、技術トレンドへの投資や対応策を提案できているか。 |
| 5. 業界課題2 | 需要減退・人材不足対応 | 工事需要の変化や深刻な人手不足（2024年問題、外国人材受入等）に対し、実効性のある解決策（歩掛管理の高度化、採用・定着策等）を示せているか。 |

# 出力形式
## 1. 総合評価
（A/B/C のランクと、その理由を簡潔に）

## 2. 項目別評価
- **全体構成**: [ランク] コメント...
- **地域性**: [ランク] コメント...
- **業界特性**: [ランク] コメント...
- **業界課題1 (GX/DX)**: [ランク] コメント...
- **業界課題2 (人材/需要)**: [ランク] コメント...

## 3. 良い点（Strengths）
- 
- 

## 4. 改善のためのアドバイス（Areas for Improvement）
- 
-
'''

# 提案書（test_proposal変数）
proposal_text = test_proposal

# 評価実行
evaluation_result = evaluate_proposal(proposal_text, evaluation_prompt)

print("=== 提案書評価結果 ===")
print(evaluation_result)
```

## Step 5: 全社一括提案書・評価生成＆CSV出力

```python
# Step 5: 全社一括提案書・評価生成＆CSV出力

results = []
for code in financial_df['コード'].unique():
    try:
        proposal = generate_proposal(code, top_k=3)
        evaluation = evaluate_proposal(proposal, evaluation_prompt)
        results.append({'コード': code, '提案書': proposal, '評価': evaluation})
    except Exception as e:
        results.append({'コード': code, '提案書': f'生成失敗: {e}', '評価': '評価不可'})

# DataFrame化・CSV出力
proposals_df = pd.DataFrame(results)
proposals_df.to_csv('提案書_評価一覧.csv', index=False)

print('全社分の提案書・評価をCSVに出力しました。')
display(proposals_df)
```
